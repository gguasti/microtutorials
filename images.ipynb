```python
# Warning:
# Color image loaded by OpenCV is in BGR mode.  But Matplotlib displays in RGB mode.  So color
# images will not be displayed correctly in Matplotlib if image is read with OpenCV.

import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('incredibles.jpg')
# b,g,r = cv2.split(img)
# img2 = cv2.merge([r,g,b])
img2 = img[:,:,::-1]
img3 = img[::-1, :, ::-1]

plt.subplot(131);plt.imshow(img) # expects distorted color
plt.subplot(132);plt.imshow(img2) # expect true color
plt.subplot(133);plt.imshow(img3) # expect flip vertical
plt.show()


# this freezes in my case - common issue!
# cv2.imshow('bgr image',img) # expects true color
# cv2.imshow('rgb image',img2) # expects distorted color
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
